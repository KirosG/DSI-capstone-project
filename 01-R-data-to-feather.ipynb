* .csv: `data.table`
* .rds, .rda, .R, .RData: `base`
* .sav, .dta: `haven`
* .xls, .xlsx: `readxl`, `openxlsx`

### 2\. Exporting Select Dataframes to CSV or Other Preferred Format


One idea is the [following code recipe](http://stat.ethz.ch/R-manual/R-devel/library/utils/html/write.table.html) for CSV files:

> ```R
write.table(x, file = "", append = FALSE, quote = TRUE, sep = " ",
            eol = "\n", na = "NA", dec = ".", row.names = TRUE,
            col.names = TRUE, qmethod = c("escape", "double"),
            fileEncoding = "")
write.csv(...)
write.csv2(...)
```
